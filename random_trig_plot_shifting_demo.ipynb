```python
import matplotlib.pyplot as plt
import matplotlib.ticker as tck
import numpy as np
import random
```

```python
f,ax=plt.subplots(figsize=(10,5))
ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
ax.xaxis.set_major_locator(tck.MultipleLocator(base=1.0))

# UD: up-down shifting
# LR: left-right shifting
UD = False
LR = False
# UD = True
# LR = True

UD_shift = 0
LR_shift = 0
sym1 = ''
sym2 = ''

x = np.array(range(10))
lower = -1*random.choice(x)
upper = abs(random.choice(x))

print(lower,upper)

funcs = ['sin', 'cos', 'tan']
# funcs = ['sec', 'csc', 'cot']
# funcs = ['sin', 'cos', 'tan', 'sec', 'csc', 'cot']


func_name = random.choice(funcs)
x_range = np.linspace(lower-30,upper+30,800)


if func_name == 'tan' or 'cot':
  ax.set_ylim(-10,10)


func = {'sin': np.sin, 'cos': np.cos, 'tan': np.tan, 'sec':np.cos, 'csc': np.sin, 'cot': np.tan}[func_name]



if LR:
  LR_shift = random.choice(np.array([-1.5,-1,-.5,0,.5,1,1.5]))
  if LR_shift >= 0:
    sym1 = '+'

if UD:
  UD_shift = random.choice(range(-5,6))
  if UD_shift >=0:
    sym2 = '+'

y_range = func(x_range+LR_shift) + UD_shift
if func_name in ['sec','csc','cot']:
  y_range = 1/func(x_range+LR_shift) + UD_shift

ax.axvline(x=0, c="black", label="x=0")
ax.axhline(y=0, c="black", label="y=0")
ax.plot(x_range,y_range)
ax.set_xlim(lower,upper)

print(f'{func_name}(θ{sym1}{LR_shift}π){sym2}{UD_shift}')
print(func)
print(f'UD: {UD}, {UD_shift}')
print(f'LR: {LR}, {LR_shift}')
```
